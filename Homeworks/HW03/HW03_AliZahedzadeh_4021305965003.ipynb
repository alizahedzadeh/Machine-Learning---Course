## <font color='#D61E85' size='+3'>**Q1:**</font> <font size='+2'> **Multivariate least squares** </font>

In a regression task, suppose that instead of trying to predict a single output, we have a training set with multiple outputs for each example:

$$\{(x^{(i)}, y^{(i)}), i=1, ...,m\}, x^{(i)} ‚àà \mathbb{R}^n, y^{(i)} ‚àà \mathbb{R}^p. $$

Thus for each training example, $y^{(i)}$ is vector-valued, with $p$ entries. We wish to use a linear model to predict the outputs, as in least squares, by specifying the parameter matrix $ùöØ$ in

$$y = ùöØ^T x,$$

where $ùöØ ‚àà \mathbb{R}^{n√óp}$.

<font size='+1'>**(a)**</font> The cost function for this case is

$$J(ùöØ) = \frac{1}{2} \sum_{i=1}^m \sum_{j=1}^p \Big((ùöØ^Tx^{(i)})_j - y^{(i)}_j\Big)^2.$$

Write $J(ùöØ)$ in matrix-vector notation (i.e., without using any summations). [Hint: Start with the $m √ó n$ design matrix

$$
X =
\begin{bmatrix}
\overline{\space\space\space\space\space\space} & (x^{(1)})^T & \overline{\space\space\space\space\space\space}\\

\end{bmatrix}
$$

and the $m √ó p$ target matrix

<font size='+1'>**(b)**</font> Find the closed form solution for $ùöØ$ which minimizes $J(ùöØ)$. This is the equivalent to the normal equations for the multivariate case.

<font size='+1'>**(c)**</font> Suppose instead of considering the multivariate vectors $y^{(i)}$ all at once, we instead compute each variable $y^{(i)}_j$ separately for each $j = 1,\dots,p$. In this case, we have a $p$ individual linear models, of the form

<font color='#8FCF26' size='+2'>**A1:**</font> Part (a), (b), and (c)

### **Answer -> Part (A)**

\begin{align}
J(\Theta) = \frac{1}{2} \sum_{i=1}^m \sum_{j=1}^p \left((\Theta^T x^{(i)})_j - y^{(i)}_j\right)^2
\end{align}



The linear model $ \ y = \Theta^T x \ $  can be written as $ \ X\Theta \ $, where $ \ \Theta \ $ is a matrix of parameters.


\begin{align}
J(\Theta) = \frac{1}{2} \sum_{i=1}^{m} \sum_{j=1}^{p} (X\Theta - Y)^2_{ij}\\
\end{align}



\begin{align}
J(\Theta) = \frac{1}{2} \sum_{i=1}^{n} \left( (X\Theta - Y)^T(X\Theta - Y) \right)_{ii}
\end{align}

\begin{align}
J(\Theta) = \frac{1}{2} \text{tr} \left( (X\Theta - Y)^T (X\Theta - Y) \right)
\end{align}

### **Answer -> Part (B)**

We need to take the gradient of $ J (Œò) $ with respect to $ Œò $.

\begin{equation}
\nabla_{\Theta} J(\Theta) = \nabla_{\Theta} [\frac{1}{2} \text{tr} \left( (X\Theta - Y)^T (X\Theta - Y) \right)]
\end{equation}

\begin{equation}
 = \nabla_{\Theta} [\frac{1}{2} \text{tr} \left( \Theta^T X^T X \Theta - \Theta^T X^T Y - Y^T X \Theta - Y^T Y \right)]
\end{equation}

\begin{equation}
 = \frac{1}{2} \left( \nabla_{\Theta} \text{tr}(\Theta^T X^T X \Theta) - \nabla_{\Theta} \text{tr}(\Theta^T X^T Y) - \nabla_{\Theta} \text{tr}(Y^T X \Theta) + \nabla_{\Theta} \text{tr}(Y^T Y) \right)
\end{equation}

\begin{equation}
 = \frac{1}{2} \left( \nabla_{\Theta} \text{tr}(\Theta^T X^T X \Theta) - 2 \nabla_{\Theta} \text{tr}(Y^T X \Theta) + \nabla_{\Theta} \text{tr}(Y^T Y) \right)
\end{equation}

\begin{equation}
= \frac{1}{2} \left( X^T X \Theta + X^T X \Theta - 2X^T Y \right)
\end{equation}




Final Result :


\begin{equation}
J(\Theta) = X^T X \Theta - X^T Y
\end{equation}

Set it to Zero :

\begin{align}
X^T(X\Theta - Y) &= 0 \\
\end{align}


Solving for $ \Theta $ :


\begin{equation}
X^T X \Theta = X^T Y
\end{equation}

\begin{equation}
\Theta = (X^TX)^{-1}X^TY
\end{equation}

We have set of vectors

\begin{equation}
\mathbf{y}_j = \begin{bmatrix} y_j^{(1)} \\ y_j^{(2)} \\ \vdots \\ y_j^{(m)} \end{bmatrix}, \quad j = 1, 2, \ldots, p.
\end{equation}

Then each linear model can be solved by the least squares solution

\begin{equation}
\theta_j = (X^T X)^{-1} X^T \mathbf{y}_j
\end{equation}


If we line up our $ Œ∏j $ , we see that we have the following equation:

\begin{equation}
\begin{bmatrix} \theta_1 & \theta_2 & \dots & \theta_p \end{bmatrix} = (X^T X)^{-1} X^T \begin{bmatrix} \mathbf{y}_1 & \mathbf{y}_2 & \dots & \mathbf{y}_p \end{bmatrix}
\end{equation}


\begin{equation}

we can see that the parameters obtained in the multivariate solution are a concatenation of the parameters obtained independently for each variable in the univariate approach. In other words, the multivariate solution considers all variables simultaneously and provides a joint solution, while the univariate approach gives separate solutions for each variable independently.

**Normal Equation and Intercept**

The ```Normal Equation``` is a method for finding the optimal parameters (coefficients) of a linear regression model. It provides a closed-form solution to the linear regression optimization problem.

```Intercept``` in linear regression refers to the coefficient associated with the constant term or bias term in the model

```add_intercept```

Function that adds a column of ones to a given matrix. This function is commonly used to include the intercept term in linear regression models.

```normal_equation```

The normal_equation function computes the normal equation solution including the intercept

````python
def add_intercept(x):
    # Add a column of ones to the design matrix for the intercept term
    return np.c_[np.ones(x.shape[0]), x]

def normal_equation(x, y):
    # Compute the normal equation solution
    theta = np.linalg.inv(x.T @ x) @ x.T @ y
    return theta
````

**Split data (Train , Test , Validation)**

````python
# Define features and targets variable
Y = housing_data[['crim', 'tax']]
X = housing_data.drop(['crim', 'tax'], axis=1)
````

````python
# Add an intercept term to the feature matrices
x_train = add_intercept(x_train)
x_val = add_intercept(x_val)
x_test = add_intercept(x_test)

# Compute theta using the normal equations for the training set
thetha_normal_equation = normal_equation(x_train, y_train)
thetha_normal_equation
````

**Training data Metrics**

````python
y_pred_train = x_train @ thetha_normal_equation
mse = mean_squared_error(y_train, y_pred_train)
r2 = r2_score(y_train, y_pred_train)
mae = mean_absolute_error(y_train, y_pred_train)
mape = mean_absolute_percentage_error(y_train, y_pred_train)

print("MSE :", mse)
print("R2 Score :", r2)
print("MAE :", mae)
print("MAPE :", mape)
````

**Validation data Metrics**

````python
y_pred_validation = x_val @ thetha_normal_equation
mse = mean_squared_error(y_val, y_pred_validation)
r2 = r2_score(y_val, y_pred_validation)
mae = mean_absolute_error(y_val, y_pred_validation)
mape = mean_absolute_percentage_error(y_val, y_pred_validation)

print("MSE :", mse)
print("R2 Score :", r2)
print("MAE :", mae)
print("MAPE :", mape)
````

### **Answer -> Part (A)**

```Regularization``` is a technique used to reduce errors by fitting the function appropriately on the given training set and avoiding overfitting. The commonly used regularization techniques are : 

- Lasso Regularization ‚Äì L1 Regularization
- Ridge Regularization ‚Äì L2 Regularization

**Lasso Regularization**

A regression model which uses the L1 Regularization technique is called LASSO(Least Absolute Shrinkage and Selection Operator) regression. Lasso Regression adds the ``absolute value of magnitude`` of the coefficient as a penalty term to the loss function(L). Lasso regression also helps us achieve feature selection by penalizing the weights to approximately equal to zero if that feature does not serve any purpose in the model

\begin{align}
J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \left( h_\theta(x^{(i)}) - y^{(i)} \right)^2 + \lambda \sum_{j=1}^{n} \left| \theta_j \right| \
\end{align}

**Ridge Regression**

A regression model that uses the L2 regularization technique is called Ridge regression. Ridge regression adds the ```squared magnitude``` of the coefficient as a penalty term to the loss function.

\begin{align}
\ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \left( h_\theta(x^{(i)}) - y^{(i)} \right)^2 + \lambda \sum_{j=1}^{n} \theta_j^2 \
\end{align}

1. **$ l^1 $ Regularization:**

   The $ l^1 $ regularization term forms a diamond-shaped constraint in the parameter space.
   This diamond intersects with the axes, creating corners and  These corners correspond to situations where one or more weights $ \beta_j $ become exactly zero.
   The geometric interpretation of $ l^1 $ regularization emphasizes its ability to promote sparsity by driving some of the weights to zero. The corners of the diamond represent sparse solutions where certain features are effectively ignored.

2. **$ l^2 $ Regularization:**

   The $ l^2 $ regularization term forms a circular constraint in the parameter space.
   Unlike $ l^1 $, the circular constraint doesn't have corners, and as a result, weights are less likely to be exactly zero.
   The circular shape promotes a more distributed impact on the weights. 
   $ l^2 $ regularization is often seen as providing a more balanced regularization effect across all features.

The above picture showed a visual representation of how $ l^1 $ and $ l^2 $ regularization influence the parameter space. $ l^1 $ regularization, forming a diamond shape, encourages sparsity by driving certain weights to exactly zero, while $ l^2 $ regularization, with its circular shape, provides a more distributed impact across all weights without enforcing sparsity.

### **Answer -> Part (B)**

The choice between the two regularization methods depends on the desired characteristics of the model and the specific requirements of the problem at hand in the domain of private law and related research areas.

The main difference between them is that L1 regularization helps in feature selection by eliminating the features that are not important, but l2 is helpful when the number of feature points are large in number.

````python
mean = 0
std = 1
noise = np.random.normal(mean, std, Y.shape) * 0.1
noisy_y = Y + noise
````

- Determine the best and worst degree in terms of the total cost.

````python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Define variables
best_degree = None
worst_degree = None
best_cost = float('inf')
worst_cost = float('-inf')
all_costs = []
all_degrees = []
all_predictions = []
all_models = []

for degree in range(1, 16):  
    poly = PolynomialFeatures(degree=degree)
    x_poly = poly.fit_transform(X.reshape(-1, 1))

    #define linear model
    model = LinearRegression()
    model.fit(x_poly, noisy_y)
    all_models.append(model)

    #prediction
    noisy_y_pred = model.predict(x_poly)
    all_predictions.append(noisy_y_pred)

    #calculate cost
    cost = mean_squared_error(noisy_y, noisy_y_pred)
    all_costs.append(cost)
    all_degrees.append(degree)
    

    if cost < best_cost:
        best_cost = cost
        best_degree = degree

    if cost > worst_cost:
        worst_cost = cost
        worst_degree = degree
````

## <font color='#D61E85' size='+3'>**Q5:**</font> <font size='+2'> **Logistic Regression** </font>

In this question you need to implement logistic regression from scratch. Then you will use it and try to reproduce the results of previous question.

````python
class MyLogisticRegression:

    def __init__(self, lr, num_iter, random_state=42):
        """
        lr: learning rate
        num_iter: number of iterations over the training data
        random_state: random state to initialize random weights, don't change the 42
        """
        self.lr = lr
        self.num_iter = num_iter
        self.random_state = random_state

    def fit(self, X, y):
        """
        Parameters
        -------
        X : Training data, shape = [n_examples, n_features]
        y : Target values, shape = [n_examples]

        return
        -------
        self : Instance of LogisticRegression
        """

        # initialize weights using the normal distribution (mean=0, std=0.01)
        # use weights[0] as bias and weights[1:] as non-bias weights
        my_range = np.random.default_rng(self.random_state)
        self.weights = my_range.normal(0, 0.01, X.shape[1] + 1)

        self.losses = list()

        for _ in range(self.num_iter):
            f_x = np.dot(X, self.weights[1:]) + self.weights[0]
            output = self.activation(f_x)
            
            # calculate the error and update the model parameters
            error = y - output
            self.weights[1:] += self.lr * np.dot(error, X) / X.shape[0]
            self.weights[0] += self.lr * np.sum(error) / X.shape[0]

            # calculate the cost
            loss = self.calculate_loss(y, output)
            self.losses.append(loss)

        return self

    def activation(self, f_x):
        # Compute logistic sigmoid activation
        return 1 / (1 + np.exp(-f_x))

    def predict(self, X):
        # Return class label
        return np.round(self.activation(np.dot(X, self.weights[1:]) + self.weights[0]))

    def calculate_loss(self, y, output):
        # Calculate binary cross-entropy loss
        return -np.mean(y * np.log(output) + (1 - y) * np.log(1 - output))
````

**Explaination**

**```__init__```**:

It's a method.

It's the class cunstructor (A constructor is a unique function that gets called automatically when an object is created of a class. The main purpose of a constructor is to initialize or assign values to the data members of that class)

for example in our class it's set ```lr```,```num_iter``` and ```random_state```


**```fit ```**:

It's a method.

This method initializes the ```weights``` using a normal distribution with a mean of 0 and a standard deviation of 0.01. The bias is the first element of ```self.weights```, and the non-bias weights are the rest.

It iteratively updates the weights by computing the logistic sigmoid activation (```output```) and adjusting the weights based on the error between the predicted and true values.

The loss for each iteration is calculated using the binary cross-entropy loss function.

The loss values are stored in the ```self.losses``` list.


**```activation ```**:

It's a method.

`f_x` : The linear combination of input features and weights.

Computes the logistic sigmoid activation for the given linear combination using the formula $\frac{1}{1 + e^{-f_x}}$ .


**```predict```**:

It's a method.

After training, we can use the model to predict new data. The ```predict``` method takes some input features, calculates their weighted sum, applies the sigmoid activation, and rounds the result to 0 or 1, making a prediction.

**```predict```**:

During training, we want to know how well our model is doing. The loss is a measure of how far off our predictions are from the actual outcomes. The lower the loss, the better our model is at making predictions.
